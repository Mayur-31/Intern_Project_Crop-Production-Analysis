# BiVarate Analysis

# State-Wise Production status

```python
plt.figure(figsize= (15,15))
sns.barplot(x=df['Production'],y= df["State"], orient='h')
```

# Kerala is top state when we look at the quantum of Production for last 19.years.

# Crop wise Production status

```python
plt.figure(figsize= (15,15))
sns.barplot(x=df['Production'],y= df["Crop"], orient='h')
```

# Yearwise Production Status

```python
plt.tick_params(labelsize=10)
df.groupby("Crop_Year")["Production"].agg("sum").plot.bar()
```

# Season wise Production Status:

```python
#Season vs Production
season = df.groupby(by='Season')['Production'].sum().reset_index().sort_values(by='Production', ascending=False).head(10)
season
fig, ax = plt.subplots(figsize=(15,10))
sns.barplot(season.Season, season.Production,errwidth=0)
plt.yscale('log')
plt.title('Seasonal Crops vs Production')
```

# Crop wise Production plot describing production values for all crop types.

```python
plt.figure(figsize=(15,8))
plt.tick_params(labelsize=10)
df.groupby("Crop")["Production"].agg("count").plot.bar()
plt.show()
```
